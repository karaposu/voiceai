```python
from imputemodule import ImputeModule, WhatToRetain, EntityToImpute
from imputemodule.core.config import get_digikey_fast_only_config, FastPathConfig
from imputemodule.core.config import PipelineConfig, get_development_config
```

```python
schema = [
        WhatToRetain(name="component_taxonomy", desc="Type/Taxonomy/Category of electronic component"),
        WhatToRetain(name="voltage_rating", desc="Maximum voltage rating"),
        WhatToRetain(name="package_type", desc="Physical package type")
    ]

entity = EntityToImpute(name="BAV99")
```

```python
config = get_development_config()
imputemodule = ImputeModule(config)
```

```python
impute_op = await imputemodule.run(
        entity=entity,  
        schema=schema,
        max_urls=5,
        enable_streaming=True,
        capture_detailed_metrics=True
    )
    


    
```
